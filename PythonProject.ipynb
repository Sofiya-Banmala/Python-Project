```python
df_salary_job_contry = df.pivot_table(values='salary_year_avg', index= 'job_country', columns='job_title_short', aggfunc='median',fill_value=0)
```

```python
df_salary_job_contry.head()
```

```python
df_salary_job_contry.index
```

```python
top_countries = df['job_country'].value_counts().sort_values(ascending = False).head(10) # top 10 countries by count

# record count of job country and then sort by descending and top 10 countries with most job
```

```python
top10_countries = df['job_country'].value_counts().sort_values(ascending = False).head(10).index # top 10 countries by count
```

```python
top_countries
```

```python
top10_contries= df_salary_job_contry.loc[top10_countries]
```

```python
job_titles = ['Data Analyst','Data Engineer','Data Scientist']
```

```python
df_job_country_salary = top10_contries[job_titles]
```

```python
ax = df_job_country_salary.plot(kind = 'bar', figsize=(12,6))
ax.legend(loc='upper left')
plt.title("Job Posting across top 10 Countries")
plt.ylabel("Median Salary")
plt.tight_layout()
plt.show()

# Insights -- India and Germany has highest median salary for Data Engineer job titles, where as Netherlands has highest median salary for Data Scientist
# The insights tells us that candidate needs to switch there jobs in respective countries to earn more income.
```

```python
import ast

df_clean = df[df['job_skills'].notna()] # remove null or Nan or none case
df_clean['job_skills'] = df_clean['job_skills'].apply(ast.literal_eval) # convert to list
```

```python
df_exploded = df_clean.explode('job_skills') # explode
```

```python
df_exploded.head()
```

```python
df_exploded['job_skills'].value_counts().sort_values(ascending = False).head(10)
```

```python
df_exploded['job_skills'].value_counts().sort_values(ascending = False).head(10).plot(kind = 'barh')
```

```python
skill_counts = df_exploded.groupby(['job_title_short','job_skills']).size()
```

```python
skill_counts
```

```python
df_skill_counts = skill_counts.reset_index(name = 'skill_counts')
```

```python
df_skill_counts = df_skill_counts.sort_values(by=  'skill_counts',ascending=False)
```

```python
df_skill_counts
```

```python
df_DA_skillset.plot(kind = 'barh', x='job_skills', y='skill_counts')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 3, figsize=(18, 8), constrained_layout=True)

# Plotting with customizations
df_DA_skillset.plot(kind='barh', x='job_skills', y='skill_counts', ax=ax[0], color='skyblue', legend=False)
ax[0].set_title('Data Analyst Skills')
ax[0].invert_yaxis()  # Highest at top
ax[0].set_xlabel('Count')
ax[0].set_ylabel('Skills')

df_DE_skillset.plot(kind='barh', x='job_skills', y='skill_counts', ax=ax[1], color='lightgreen', legend=False)
ax[1].set_title('Data Engineer Skills')
ax[1].invert_yaxis()
ax[1].set_xlabel('Count')
ax[1].set_ylabel('')

df_DS_skillset.plot(kind='barh', x='job_skills', y='skill_counts', ax=ax[2], color='salmon', legend=False)
ax[2].set_title('Data Scientist Skills')
ax[2].invert_yaxis()
ax[2].set_xlabel('Count')
ax[2].set_ylabel('')

# Optional: Adjust font size if labels are long
for a in ax:
    a.tick_params(axis='y', labelsize=9)

plt.suptitle("Top Skills in Data Roles", fontsize=16)
plt.show()
```

```python
skills_stats = df_exploded.groupby('job_skills').agg(skill_count = ('job_skills','count'),median_salary=('salary_year_avg','median'))
skills_Statistics = skills_stats.sort_values(by='skill_count',ascending=False).head(10)
# skill count = column name , median_salary = column name
```

```python
skills_Statistics.plot(kind= 'scatter', x='skill_count', y='median_salary')
plt.xlabel("Job Posting")
plt.ylabel("Median Salary")

for i, text in enumerate(skills_Statistics.index):
  plt.text(skills_Statistics['skill_count'].iloc[i],skills_Statistics['median_salary'].iloc[i],text)
  plt.tight_layout()
```
